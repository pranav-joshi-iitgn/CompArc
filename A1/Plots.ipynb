#### Libraries and files

```python
import pandas as pd
from numpy import *
from matplotlib.pyplot import *
files = [
    "Q2FloatPy.csv",
    "Q2IntPy.csv",
    "Q2DoubleC.csv",
    "Q2IntC.csv",
]
Titles = [
    "Python float",
    "Python int",
    "C double",
    "C int",
]
```

### Individual variations

```python
for i,file in enumerate(files):
    df = pd.read_csv(file)
    figure()
    plot(df["N"],df["cpu_time"],'ro')
    title(Titles[i] + " Program time (s)")
    xticks(df["N"])
    xlabel("$ N $")
    ylabel(r"$ T_\text{program} $")
    X = array([[1,x,x**2,x**3] for x in df["N"].values])
    y = df["cpu_time"].values
    pseudo_inv = dot(linalg.inv(X.T @ X) @ X.T , y)
    a,b,c,d = pseudo_inv
    x = linspace(0,max(df["N"].values),1000)
    ypred = a + b*x + c*x**2 + d*x**3
    plot(x,ypred,'b')
    legend(["data",
        r"$ T_\text{program} "+f"= {round(a,2)} + ({round(b,2)})N + ({round(c,2)})N^2 + ({round(d,2)})N^3$"])
    figure()
    plot(df["N"],df["sys_time"],'r-o')
    title(Titles[i] + " System time (s)")
    xticks(df["N"])
    xlabel("$ N $")
    ylabel(r"$ T_\text{system} $")
    
```
